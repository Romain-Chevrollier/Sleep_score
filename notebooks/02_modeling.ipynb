```python
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
import numpy as np
import mlflow
import os 
```

```python
X_train = np.load("../data/X_train.npy")
X_test = np.load("../data/X_test.npy")
y_train = np.load("../data/y_train.npy")
y_test = np.load("../data/y_test.npy")
```

```python
DummyRegressor = DummyRegressor()
LinearRegression = LinearRegression()
```

```python
dummy_model = DummyRegressor.fit(X_train,y_train)
dummy_pred = dummy_model.predict(X_test)

print(r2_score(y_test, dummy_pred))
rmse = np.sqrt(mean_squared_error(y_test, dummy_pred))
print(rmse)
```

```python
linear_model = LinearRegression.fit(X_train,y_train)
linear_pred = linear_model.predict(X_test)

r2 = r2_score(y_test, linear_pred)
print(r2)
rmse = np.sqrt(mean_squared_error(y_test, linear_pred))
print(rmse)
```

```python
mlflow.set_tracking_uri("file:" + os.path.abspath("../mlruns"))

mlflow.set_experiment("sleep-fatigue-prediction")

with mlflow.start_run(run_name="linear_regression"):
    # 1. Logger les paramètres du modèle
    mlflow.log_param("model_type", "LinearRegression")
    
    # 2. Logger les métriques
    mlflow.log_metric("rmse", rmse)
    mlflow.log_metric("r2", r2)
    
    # 3. Logger le modèle
    mlflow.sklearn.log_model(linear_model, "model")
```

```python
with mlflow.start_run(run_name="random_forest"):
    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    mlflow.log_param("model_type", "RandomForest")
    mlflow.log_metric("rmse", rmse)
    mlflow.log_metric("r2", r2)
    mlflow.sklearn.log_model(model, "model")
    
    print(f"RMSE: {rmse:.4f} | R²: {r2:.4f}")
```

```python
with mlflow.start_run(run_name="xgboost"):
    model = XGBRegressor(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    mlflow.log_param("model_type", "xgboost")
    mlflow.log_metric("rmse", rmse)
    mlflow.log_metric("r2", r2)
    mlflow.sklearn.log_model(model, "model")
    
    print(f"RMSE: {rmse:.4f} | R²: {r2:.4f}")
```

```python
with mlflow.start_run(run_name="lightgbm"):
    model = LGBMRegressor(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    mlflow.log_param("model_type", "lightgbm")
    mlflow.log_metric("rmse", rmse)
    mlflow.log_metric("r2", r2)
    mlflow.sklearn.log_model(model, "model")
    
    print(f"RMSE: {rmse:.4f} | R²: {r2:.4f}")
```
